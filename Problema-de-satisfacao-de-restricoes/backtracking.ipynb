```python
class Grafo():
  
  def __init__(self, qVertices):
    self.V = qVertices
    self.grafo = [[0 for column in range(qVertices)] for row in range(qVertices)]
  
  # Funcao para verificar se uma determinada coloracao e possivel para um vertice
  def ehSeguro(self, v, atribuicaoCores, c):
    for i in range(self.V):
      if self.grafo[v][i] == 1 and atribuicaoCores[i] == c:
        return False
    
    return True



  # Recursao para o problema de coloracao
  def coloracaoUtil(self, cores, atribuicaoCores, v):
    if v == self.V:
      return True
    
    for c in cores:
      if self.ehSeguro(v, atribuicaoCores, c):
        atribuicaoCores[v] = c
        
        if self.coloracaoUtil(cores, atribuicaoCores, v + 1):
          return True
        
        atribuicaoCores[v] = None

  def coloracaoGrafo(self, m):
    atribuicaoCores = [None] * self.V

    if self.coloracaoUtil(m, atribuicaoCores, 0) == None:
      return "Nao existe solucao"

    #Impressao da solucao
    print("A solucao se da por meio da seguinte coloracao:")
    for c in atribuicaoCores:
      print (c,end=' ')
    
```

```python
g = Grafo(7)
g.grafo = [[0, 1, 0, 1, 0, 0, 1], #Western Australia
           [1, 0, 1, 1, 0, 0, 1], #Northen 
           [0, 1, 0, 1, 1, 0, 1], #Queensland
           [1, 1, 1, 0, 1, 1, 1], #South
           [0, 0, 1, 1, 0, 1, 1], #New South
           [0, 0, 0, 1, 1, 0, 1], #Victoria
           [1, 1, 1, 1, 1, 1, 0], #Tasmania
          ]
m = ["vermelho", "verde", "azul", "rosa"]
g.coloracaoGrafo(m)
```
